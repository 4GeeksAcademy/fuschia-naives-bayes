```python
#def preprocess_text(review):
    # Remove non-alphabetic characters and convert to lower case
    #return ''.join([char.lower() for char in review if char.isalpha() or char.isspace()])

import re


def preprocess_text(review):
    # Remove non-alphabetic characters and convert to lower case
    review = re.sub(r'[^A-Za-z\s]', '', review)
    review = review.lower()
    return review

# Apply preprocessing
X_train = X_train.apply(preprocess_text)
df['review'] = df['review'].apply(preprocess_text)
df.tail(20)
```

```python

#transform the text into a word count matrix
# Vectorize text data with binary (boolean) representation
vec_model = CountVectorizer(stop_words='english', binary=True, strip_accents= 'unicode')
X_train_count = vec_model.fit_transform(df['review']).toarray()

# Convert the array to a DataFrame with proper column names
X_train_count_df = pd.DataFrame(X_train_count, columns=vec_model.get_feature_names_out())

# Display the resulting DataFrame
print("Vectorized DataFrame:")
print(X_train_count_df)
```
